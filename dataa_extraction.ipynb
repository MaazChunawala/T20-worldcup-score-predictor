```python
import pandas as pd
from tqdm import tqdm
from yaml import safe_load
import os

# Initialize an empty list to store individual DataFrames
df_list = []
counter = 1

# Iterate over all files
for file in tqdm(filenames):
    # Skip non-YAML/JSON files
    if not file.endswith(('.yaml', '.yml', '.json')):
        continue

    try:
        # Attempt to open the file with UTF-8 encoding
        with open(file, 'r', encoding='utf-8') as f:
            # Load the JSON/YAML content using safe_load and normalize it into a DataFrame
            df = pd.json_normalize(safe_load(f))
    except (UnicodeDecodeError, yaml.YAMLError):
        # If decoding or parsing fails, try again with ISO-8859-1 encoding
        with open(file, 'r', encoding='ISO-8859-1', errors='ignore') as f:
            df = pd.json_normalize(safe_load(f))

    # Add a match_id column to the DataFrame
    df['match_id'] = counter
    # Append the DataFrame to the list
    df_list.append(df)
    counter += 1

# Concatenate all DataFrames in the list into a single DataFrame
final_df = pd.concat(df_list, ignore_index=True)

final_df
```

```python
matches = pickle.load(open('/Users/alichunawala/Desktop/cricket-score-predictor-main/dataset_level1.pkl','rb'))
matches.iloc[0]['innings'][0]['1st innings']['deliveries']
```

```python
import pandas as pd

count = 1
delivery_df = pd.DataFrame()

for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count += 1
        continue
    count += 1
    
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except KeyError:
                player_of_dismissed.append('0')
    
    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })
    
    # Use pd.concat instead of append
    delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True)

delivery_df
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis=1)
```
